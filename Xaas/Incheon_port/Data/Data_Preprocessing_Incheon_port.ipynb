```python
import pandas as pd
import numpy as np
import scipy.io as sio
import os

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
```

```python
def find_date(df_raw, yy, mon, day):
    for i in range(0,len(df_raw)):
        if df_raw.iloc[i,0].year == yy:
            if df_raw.iloc[i,0].month == mon: 
                if df_raw.iloc[i,0].day == day:
                    X = i
                    break
                else:
                    continue
    return X
```

```python
fn = './CG_1&CG_2_sig0_3_12.xlsx'
df_raw = pd.read_excel(fn)
df_raw
```

```python
# TT_2 && TT_4 초기 데이터(8월말 ~ 9월) 삭제
det_date1 = find_date(df_raw, 2022, 11, 1)
print(det_date1)

df_raw_tt = df_raw.iloc[det_date1:,:]
df_raw_tt.index = pd.RangeIndex(len(df_raw_tt)) #index reset
df_raw_tt
```

```python
# Date Label for xticks
point1 = find_date(df_raw_tt, 2022, 11, 1)
point2 = find_date(df_raw_tt, 2023, 1, 1)
point3 = find_date(df_raw_tt, 2023, 3, 1)
point4 = find_date(df_raw_tt, 2023, 5, 1)
point5 = find_date(df_raw_tt, 2023, 7, 1)
damage_date = find_date(df_raw_tt, 2023, 10, 1)
```

```python
plt.figure(figsize=(12, 9))

for i in range(1,10):

  plt.subplot(3,3,i)
  plt.scatter(df_raw_tt.iloc[:,0],df_raw_tt.iloc[:,i], s=0.1)
  plt.xlabel('Date')
  plt.ylabel(df_raw_tt.columns[i])
  plt.xticks([df_raw_tt.iloc[point1, 0], df_raw_tt.iloc[point3, 0], df_raw_tt.iloc[point5, 0]])
  plt.grid(True, linestyle=':', color='gray')

plt.tight_layout()

plt.show()
```

```python
# df_raw_tt 행렬의 행 개수를 확인하여 행렬 생성
num_rows = df_raw_tt.shape[0]
df_label0 = pd.DataFrame(np.zeros((num_rows, 1)), columns=['label'])
df_label0.iloc[damage_date:, :] = 1

print(df_label0)
```
