#### **Import Libraries**

```python
import pandas as pd
import numpy as np
from scipy.io import loadmat

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
```

#### **Load Preprocessed data**

```python
# displacement data
fn_disp = 'Data_2nd_Exp.mat'
dispdata = loadmat(fn_disp)
```

```python
# Dynamic data
# !pip install mat73
import mat73

fn_PSD = 'PSD_result_2ndExp.mat'
matdata = mat73.loadmat(fn_PSD)
```

```python
for data  in matdata:
    print(data)
```

```python
plt.plot(matdata['pf'])
plt.show()
```

```python
n_win = 60000
plt.plot(matdata['DIndex'][n_win-1:])
plt.show()
```

```python
df_label = pd.DataFrame(matdata['DIndex'][n_win-1:])
df_label
```

```python
df_PSD = pd.DataFrame(np.log(matdata['Pxx'].T))
df_PSD
```

```python
print(df_PSD.shape, df_label.shape)
```

```python
ind_select = np.arange(0, df_PSD.shape[0], 500)
```

```python
df_label0 = df_label.iloc[ind_select, :]
df_PSD0 = df_PSD.iloc[ind_select, :]
```

```python
print(df_label0.shape, df_PSD0.shape)
```

```python
df_PSD0 = df_PSD0.reset_index().drop(columns = 'index')
df_PSD0.to_csv('PSD_2nd_EXP_reduced.csv')

df_label0 = df_label0.reset_index().drop(columns = 'index')
df_label0.to_csv('Label_2nd_EXP_reduced.csv')
```

```python
N_pts = df_PSD0.shape[0]
IND_line_obj = np.arange(0, df_label0.shape[0])

plt.figure(figsize = (10, 4), dpi = 200)
plt.scatter(IND_line_obj, df_label0, c = IND_line_obj, cmap = 'jet')
plt.grid(ls = ':', color = 'gray')
plt.colorbar()
plt.show()
```
